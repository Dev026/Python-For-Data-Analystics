```python
def normalize(column):
    upper=column.max()
    lower=column.mean()
    y = (column - lower)/(upper - lower)
    return y
```

```python
df['Normalize_Credit_Limit']=normalize(df['Credit_Limit'])
```

```python
df['Normalize_Credit_Log_Transformed']=np.log(df['Credit_Limit'])
```

```python
fig, axes= plt.subplots(2,2, figsize=(10,7))
fig.suptitle("Before and After Transformation")
#create Plots
sns.histplot(df, x = "Credit_Limit", ax=axes[0,0])
sns.histplot(df, x = "Normalize_Credit_Limit", ax=axes[0,1])
sns.histplot(df, x = "Credit_Limit", ax=axes[1,0])
sns.histplot(df, x = "Normalize_Credit_Log_Transformed", ax=axes[1,1])
```
